```python
import pandas as pd
with open('data/lichess_kurian_2022-02-26.pgn') as file:
    file_string = file.read()
lines = file_string.split('\n')
player_name = 'kurian'
new_lines = [] #remove empty lines
colors = [] #which side has the person played
for line in lines:
    if line !='':
        new_lines.append(line)
    if player_name in line:
        colors.append(line)
# print(new_lines[0:17])
games = list()

for line in new_lines:
    if line.startswith('1.'):
        games.append(line)


game_lines = dict(games=games,color=colors)

df = pd.DataFrame(game_lines)
```

### Games By Color

```python
df['color'].value_counts().plot.pie()
```

```python
df['game_length']=df['games'].apply(lambda x:len(x.split('.')))
```

```python
df.head()
```

### Games By Length

```python
df['game_length'].value_counts().plot.pie(figsize=(10,10))
```

```python
df['opening']=df['games'].apply(lambda x: '.'.join(x.split('.')[:2:1]).replace('1.','').replace('2.',''))
```

```python
df['opening'].head()
```

### All Openings

```python
df['opening'].value_counts().plot.pie(figsize=(10,10))
```
